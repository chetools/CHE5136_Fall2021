```python
import numpy as np
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default='plotly_dark'
```

```python
N=100
x=np.linspace(0,10,N).reshape(-1,1)
csf = np.array([[3.,2.,2.],
                [4.,3.5, 5.],
                [10.,3., 6.5]])

y=np.sum(csf[:,0]*np.cos(np.expand_dims(csf[:,2],0)*x)+csf[:,1]*np.sin(np.expand_dims(csf[:,2],0)*x),1)
```

```python
f = np.arange(1.5,7,0.5)
M=f.size*2
p=np.zeros((N,M))
p[:,::2]=np.cos(np.expand_dims(f,0)*x)
p[:,1::2]=np.sin(np.expand_dims(f,0)*x)
```

```python
p[:,::2].shape
```

```python
fig = make_subplots()
fig.add_scatter(x=x.flatten(),y=y)
```

```python
pmr=(p.T @ p)/N
amr=np.zeros((M,M))
D[0,0]=1
D[:,0]=np.mean(p,0)
for m in range(1,M):
    for r in range(m-1):
        amr[m,r]=D[m,r]/D[r,r]
        D[m,r+1]=pmr[m,r+1]-np.dot(amr[r+1,:r+1],D[m,:r+1])

    Dmm = np.diagonal(D)
c = np.zeros(M)
Q=np.zeros(f.size)

N_f=3
a_values=[]
b_values=[]
w_values=[]
yy=np.copy(y)
for ii in range(N_f):
    c[0]=np.mean(yy)
    for m in range(1,M):
        c[m]=np.mean(yy*p[:,m])-np.dot(amr[m,:m],c[:m])
    g =c/Dmm

    Q=(g[::2]**2)*Dmm[::2]+(g[1::2]**2)*Dmm[1::2]
    w=f[np.argmax(Q)]
    cos,sin=np.cos(w*x).flatten(),np.sin(w*x).flatten()
    aa=(np.dot(y,cos)/np.sum(cos**2))
    bb=(np.dot(y,sin)/np.sum(sin**2))
    yy=yy-aa*cos-bb*sin
    a_values.append(aa)
    b_values.append(bb)
    w_values.append(w)
```

```python
a_values=np.array(a_values)
b_values=np.array(b_values)
w_values=np.array(w_values)
```

```python
y_fit = np.sum(a_values*np.cos(np.expand_dims(w_values,0)*x)+b_values*np.sin(np.expand_dims(w_values,0)*x),1)
```

```python
fig2 = make_subplots()
fig2.add_scatter(x=x.flatten(),y=y,mode='markers')
fig2.add_scatter(x=x.flatten(),y=y_fit,mode='lines')
```
