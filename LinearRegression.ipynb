```python
import numpy as np
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default='plotly_dark'
```

```python
x = np.linspace(0,10, 101)
y = 3*x + 1
ydata = np.random.normal(loc=y, scale=1)
```

```python
fig = make_subplots(rows=1,cols=1)
fig.add_scatter(x=x, y=ydata,mode='markers', marker_color='rgba(0,0,0,0)', marker_line_color='white',marker_line_width=0.7)
fig.update_layout(width=600,height=400)
```

```python
X=np.c_[x, np.ones_like(x)] 
```

```python
m, b = np.linalg.inv(X.T @ X) @ X.T @ ydata
```

```python
fig.add_scatter(x=x, y= m*x+b, mode='lines')
```
