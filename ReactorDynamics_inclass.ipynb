```python
import numpy as np
from scipy.integrate import solve_ivp
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default='plotly_dark'
```

```python
q = 100. #L/min
cain = 1. #mol/L
Tin = 350. #K
V = 100. #L
rho = 1000. #g/L
cp = 0.239 #J/(g K)
negHr = 5e4 #J/mol
ER = 8750 #K
k0 = 7.2e10 #1/min
UA = 5e4 #J/(min K)
Tc0 = 300. #K
ca0 = 0.5 #mol/L
T0 = 350 #K
tend = 10. #min
```

```python
def rhs(t,z):
    Tc = 305
    ca, T = z
    k=k0*np.exp(-ER/T)
    dca = q*(cain - ca)/V - k*ca
    dT = q*(Tin - T)/V + negHr*k*ca/rho/cp + UA*(Tc - T)/(rho*V*cp)
    return dca, dT
```

```python
solve_ivp(rhs,[0,tend], [ca0,T0], method='Radau', dense_output=True)
```

```python
sol = solve_ivp(rhs,[0,tend], [ca0,T0], method='Radau', dense_output=True).sol
t_plot=np.linspace(0,tend,200)
ca, T = sol(t_plot)
fig = make_subplots(rows=1, cols=2)
fig.add_scatter(x=t_plot, y=ca,row=1,col=1)
fig.add_scatter(x=t_plot,y=T, row=1,col=2)
```
