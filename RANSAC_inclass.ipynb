```python
import numpy as np
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default='plotly_dark'
```

```python
np.random.seed(360)
N_lines=5
data_std=2
x_data = np.random.uniform(-10,10, size=(1000,N_lines))
y_data = np.random.normal(loc=np.random.uniform(-10,10,size=(1,N_lines))*x_data + np.random.uniform(-50,50, size=(1,N_lines)), scale=data_std)
fig= make_subplots()
for i in range(N_lines):
    fig.add_scatter(x=x_data[:,i], y=y_data[:,i], mode='markers', marker_color='rgba(0,0,0,0)', marker_line_width=0.2,
                    marker_line_color=f'hsv({i/N_lines*360},100%,100%)')
fig.update_layout(width=800,height=600, showlegend=False)
```

```python


x_flat = np.ravel(x_data)
y_flat = np.ravel(y_data)
n_flat = x_flat.size
random_idx=np.random.permutation(np.arange(n_flat))
x_flat=x_flat[random_idx]
y_flat=y_flat[random_idx]

fig= make_subplots()

fig.add_scatter(x=x_flat, y=y_flat, mode='markers', marker_color='rgba(0,0,0,0)', marker_line_width=0.2,
                    marker_line_color='rgb(100,100,100)')
fig.update_layout(width=800,height=600, showlegend=False)
```

```python
N_cand = 100
idx1, idx2 = np.random.choice(np.arange(n_flat), size=(2,100), replace=False)
m=(y_flat[idx2]-y_flat[idx1])/(x_flat[idx2]-x_flat[idx1])
b=y_flat[idx1]-m*x_flat[idx1]
```

```python
consistent_data=np.abs(y_flat[:,None] - (m[None,:] *x_flat[:, None] + b[None,:]))<data_std
```

```python
np.sum(consistent_data,axis=0) > n_flat/n_lines/
```
