```python
import numpy as np
import tools.che as che
from scipy.optimize import root
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default='plotly_dark'
R=8.314 # J/(mol K)
```

```python
props=che.Props(['nHexane','nOctane'])
```

```python
T=70+273.15
```

```python
mf_array = np.linspace(0,1,100)  #represents x1 for bubbleP calculation or y1 for dewP calculation
bubbleP_values=[]
dewP_values=[]
for mf in mf_array:
    partial_p = np.array([mf, 1-mf])*props.Pvap(T)
    bubbleP = np.sum(partial_p)
    bubbleP_values.append(bubbleP)
    dewP_values.append(1/np.sum(np.array([mf, 1-mf])/props.Pvap(T)))
```

```python
fig=make_subplots()
fig.add_scatter(x=mf_array, y=bubbleP_values,name='bubbleP')
fig.add_scatter(x=mf_array, y=dewP_values,name='dewP')
fig.update_layout(title=f'T={T} K')
```

```python

P=101325
```

```python
def bubbleT_eq(v,x):
    T=v[0]
    return [np.sum(x*props.Pvap(T))-P]

def dewT_eq(v,y):
    T=v[0]
    return [P*np.sum(y/props.Pvap(T)) -1 ]
```

```python
mf_array = np.linspace(0,1,100)  #represents x1 for bubbleT calculation or y1 for dewT calculation
bubbleT_values=[]
dewT_values=[]
for mf in mf_array:
    bubbleT_values.append(root(bubbleT_eq, [300], args=np.array([mf,1-mf])).x[0])
    dewT_values.append(root(dewT_eq, [300], args=np.array([mf,1-mf])).x[0])
```

```python
fig2=make_subplots()
fig2.add_scatter(x=mf_array, y=bubbleT_values,name='bubbleT')
fig2.add_scatter(x=mf_array, y=dewT_values,name='dewT')
fig2.update_layout(title=f'P={P} K')
```

```python
z=np.array([0.5,0.5])
P=101325
T=370
K = props.Pvap(T)/P
```

```python
def rachford(v):
    VF = v[0]
    return [np.sum(z*(K-1)/(1+VF*(K-1)))]
```

```python
VF=root(rachford, [0.5]).x[0]
x = z/(1+VF*(K-1))
y=K*x
```

```python
print(f'liquid mole fractions (x): {x}')
print(f'vapor mole fractions (y): {y}')
print(f'Vapor mole fraction: {VF}')
```
