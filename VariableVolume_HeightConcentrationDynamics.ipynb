```python
import numpy as np
from scipy.integrate import solve_ivp
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default='plotly_dark'
```

```python
alpha = 0.08
Ac = 1.
h0 = 1.
ca0 = 0.5
tend = 200
```

```python
def rhs(t,z):
    qin = 0.1
    cain = 0.1
    h, ca = z
    dh = (qin - alpha * h**0.5)/Ac
    dca = ((qin*cain - ca*alpha*h**0.5)/Ac - ca*dh)/h
    return dh, dca
```

```python
sol=solve_ivp(rhs, (0,tend), [h0, ca0], method='Radau', dense_output=True).sol
```

```python
t_plot = np.linspace(0,tend, 200)
h, ca=sol(t_plot)
fig = make_subplots(rows=1, cols=2)
fig.add_scatter(x=t_plot, y=h, row=1, col=1, name='h')
fig.add_scatter(x=t_plot, y=ca, row=1, col=2, name='ca')
```
